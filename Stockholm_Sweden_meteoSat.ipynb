```python
import requests
import pandas as pd

# --- CONFIGURATION ---
# Stockholm, Sweden
LAT = 59.3293
LON = 18.0686

# Date Range: Full Year 2024
START_DATE = "2024-01-01"
END_DATE = "2025-01-01"

# API Endpoint (Historical Satellite Data)
url = "https://archive-api.open-meteo.com/v1/archive"

# Parameters:
# temperature_2m      = Air Temperature
# wind_speed_10m      = Wind Speed
# shortwave_radiation = GHI (Global Horizontal Irradiance)
params = {
    "latitude": LAT,
    "longitude": LON,
    "start_date": START_DATE,
    "end_date": END_DATE,
    "hourly": "temperature_2m,wind_speed_10m,shortwave_radiation",
    "timezone": "UTC"
}

print(f"Fetching Satellite (Meteosat) Data for Stockholm...")

try:
    r = requests.get(url, params=params)

    if r.status_code == 200:
        data = r.json()

        if 'hourly' in data:
            hourly = data['hourly']

            df = pd.DataFrame({
                'Time': pd.to_datetime(hourly['time']),
                'Temperature_C': hourly['temperature_2m'],
                'Wind_Speed_ms': hourly['wind_speed_10m'],
                'GHI_Wm2': hourly['shortwave_radiation'] # Solar Radiation
            })

            # Formatting
            df['Date'] = df['Time'].dt.strftime('%Y-%m-%d')
            df['Hour'] = df['Time'].dt.strftime('%H:%M')

            # Select and Reorder columns
            cols = ['Date', 'Hour', 'Temperature_C', 'Wind_Speed_ms', 'GHI_Wm2']
            df = df[cols]

            print("-" * 30)
            print(df.head())
            print("-" * 30)
            print(f"Total Rows: {len(df)}")

            filename = "Stockholm_Meteosat_2024.csv"
            df.to_csv(filename, index=False)
            print(f"Success! Data saved to {filename}")
        else:
            print("API returned 200 OK but no 'hourly' data found.")
    else:
        print(f"Error {r.status_code}: {r.text}")

except Exception as e:
    print(f"Connection failed: {e}")
```
