```python
import pandas as pd

# 1. Load your file (replace with your file name)
filename = 'csv_60.45175_22.248577_fixed_23_180_PT10M.csv'
df = pd.read_csv(filename)

# 2. Convert the time column to datetime objects
df['period_end'] = pd.to_datetime(df['period_end'])

# 3. Set the time column as the index for resampling
df.set_index('period_end', inplace=True)

# 4. Define aggregation: Mean for numbers, First for text/weather
# Identify numerical columns
numeric_cols = df.select_dtypes(include=['number']).columns
agg_dict = {col: 'mean' for col in numeric_cols}

# Add categorical columns (like weather_type) if they exist
if 'weather_type' in df.columns:
    agg_dict['weather_type'] = 'first'

# 5. Resample to Hourly ('H')
df_hourly = df.resample('H').agg(agg_dict)

# 6. Update the period identifier and clean up
df_hourly['period'] = 'PT1H'
df_hourly.reset_index(inplace=True)
df_hourly = df_hourly.round(2) # Round values for readability

# 7. Save the result
df_hourly.to_csv('hourly_data.csv', index=False)

# Display the first few rows
print(df_hourly.head())
```
